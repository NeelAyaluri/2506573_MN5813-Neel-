# Data Cleaning

**Data Cleaning & Data Wrangling**


**---Drop Irrelevant Columns**: A few columns which were not relevant for the analysis were dropped to simplify the dataset, thus improving the speed of processing.

**---Missing Value Treatment**: A few numeric columns had some missing values
 Example: Wicket type, player dismissed, city, winner and player of the match were replaced with correct names.

**---Columns Renaming**: Several columns in the datasets had to be uniformly named also done a column split.

Data cleaning and wrangling is performed to ensure that the dataset does not contain any irrelevant data. Dropping unnecessary columns reduced computational overhead, and handling missing values was important for the integrity of the dataset. Renaming of columns was necessary to avoid conflicts and to be clear while analyzing.

```python
#Data Cleaning
#Columns_to_drop = ['start_date', 'non_striker', 'other_wicket_type','other_player_dismissed', 'cricsheet_id', 'result', 'dl_applied', 'venue_y', 'umpire1', 'umpire2', 'umpire3']
#Data = Data.drop(columns = Columns_to_drop)
#Dropping the above columns as there is no data in columns.
Columns_to_drop = ['start_date', 'non_striker', 'other_wicket_type', 'other_player_dismissed', 'cricsheet_id', 'result', 'dl_applied', 'venue_y', 'umpire1', 'umpire2', 'umpire3'] 
Data = Data.drop(columns=Columns_to_drop)
Data
```

```python
Columns_to_drop = ['wides', 'noballs', 'byes', 'legbyes', 'penalty', 'cricsheet_id', 'season_y', 'team1', 'team2']
existing_columns = Data.columns.intersection(Columns_to_drop)
Data = Data.drop(columns=existing_columns)
#Dropping the above columns as well from the second dataset as there is no data in the columns.
```

```python
Data
```

```python
Data.rename(columns = {'season_x':'Season','venue_x':'Venue','striker':'Batsman'},inplace = True)
Data
#Renaming the columns
#season_X to Season, venue_x to Venue and striker to batsman for an easy lookout.
```

```python
Data.rename(columns = {'match_id':'Match ID','innings':'Innings','ball':'Ball','batting_team':'Batting Team',
                       'bowling_team':'Bowling Team','bowler':'Bowler','runs_off_bat':'Runs_off_bat','extras':'Extras','wicket_type':
                       'Wicket type','player_dismissed':'Player dismissed','city':'City','date':'Date','toss_winner':'Toss winner',
                       'toss_decision':'Toss decision','winner':'Winner','win_by_runs':'Win by runs','win_by_wickets':'Win by wickets',
                       'player_of_match':'Player of match'},inplace = True)
Data
#Renaming almost all the columns names in such a way that the starting letter should be 'capital', so that the headers look good :) 
```

```python
# Convert 'Date' column to datetime format
Data['Date'] = pd.to_datetime(Data['Date'], format='%Y/%m/%d')

# Change format to "day, month name, year" (e.g., "18 April 2008")
Data['Date'] = Data['Date'].dt.strftime('%d %B %Y')
#Changing the year format to look good and and eye catching.
```

```python
Data
```

```python
New_order = ['Match ID', 'Season', 'Date', 'Venue', 'City', 'Toss winner', 'Toss decision', 'Batting Team', 
             'Bowling Team', 'Innings', 'Ball', 'Batsman', 'Bowler', 'Runs_off_bat', 'Extras', 'Wicket type', 
             'Player dismissed', 'Winner', 'Win by runs', 'Win by wickets', 'Player of match']
Data = Data[New_order]
Data
#Re-ordering the columns for proper order and better structure.
```

```python
# Split 'Venue' into 'Stadium' and 'Location'
Data[['Stadium', 'Location']] = Data['Venue'].str.split(', ', n=1, expand=True)

# For rows where 'Location' is missing (NaN), assign 'Unknown'
Data['Location'].fillna('Unknown', inplace=True)
# We did a split for the venue column to stadium and location as some of the columns in the venues have both and some have only the stadium name.
```

```python
Data
```

```python
Columns_to_drop = ['Venue', 'Location']
Data = Data.drop(columns=Columns_to_drop)
Data
# As we had a split of stadium and location in the above and we already have city, we are dropping the venue and location
```

```python
New_order = ['Match ID', 'Season', 'Date', 'Stadium', 'City', 'Toss winner', 'Toss decision', 'Batting Team', 
             'Bowling Team', 'Innings', 'Ball', 'Batsman', 'Bowler', 'Runs_off_bat', 'Extras', 'Wicket type', 
             'Player dismissed', 'Winner', 'Win by runs', 'Win by wickets', 'Player of match']
Data = Data[New_order]
Data
# As we did changes in renaming the columns, again we are doing the column re-order for better order and proper structure.
```

```python
Data.dtypes
#identified different types of data as it will help in further analyses
```

```python
missing_values_by_column = Data.isnull().sum()
missing_values_by_column
#identified the number of missing values. It will help in analysing the data after removing the missing values
```

```python
# Wicket type and player dismissed columns were filled only when bowler took wicket and dismissed the batsman. 
# So updating the wicket type and player dismissed for other deliveries as None.

Columns_to_fill= ['Wicket type', 'Player dismissed']
Data.loc[:, Columns_to_fill] = Data[Columns_to_fill].fillna('None')
Data
```

```python
Columns_to_fill= ['City', 'Winner', 'Player of match']
Data.loc[:, Columns_to_fill] = Data[Columns_to_fill].fillna('Unknown')
Data
# City, winner and Player of match columns have missing data. Hence updated as unknown.
```

# Data Analysis


**Method Used:**

**---Data Visualization:** The visualizations are carried out using Matplotlib and Seaborn in the form of histograms, bar charts, and scatter plots.

**---Comparison Analysis:** Performances were compared among the teams, venues, and players to find out the trends.

**---Trend Analysis:** Performance was analyzed for trends like win percentage, high-scoring venues, performances of key players.

Data analysis allows the discovery of hidden patterns, correlations, and outliers in data. The visualization will thus enable easier interpretation of complex data and permit the effective communication of results. Descriptive statistics give a very good starting point for deeper analysis by summarizing the major characteristics of data.

```python
# Grouping by season and winner to count the number of wins per team each season
team_performance = ( Data.groupby(['Season', 'Winner']).size()
                    .reset_index(name = 'Wins')
                    .sort_values(by = 'Wins', ascending = False)
                   )
```

```python
team_performance
```

```python
# Data Analysis: Top Batting Teams

# Calculating total runs scored by each team
top_batting_teams = Data.groupby('Batting Team')['Runs_off_bat'].sum().sort_values(ascending=False)

print(top_batting_teams)

# Plotting top batting teams
plt.figure(figsize=(12, 6))
top_batting_teams.plot(kind='barh', color='skyblue')
plt.title('Top Batting Teams (Total Runs)')
plt.xlabel('Total Runs')
plt.ylabel('Teams')
plt.xticks(rotation=45)
plt.show()

# From the table we can see the total number of wins of a team and also we can see the same in the form of a visual representation as wee. 
#It clearly shows Mumbai Indians has won highest number of matches and they are at first place.
```

```python
# Data Analysis: Top Bowling Teams

# Counting the number of wickets taken by each bowling team
top_bowling_teams = ( Data[Data['Wicket type'].notna()].groupby('Bowling Team')
                     .size()
                     .sort_values(ascending=False)
                    )

print(top_bowling_teams)

# Plotting top bowling teams
plt.figure(figsize=(12, 6))
top_bowling_teams.plot(kind='barh', color='orange')
plt.title('Top Bowling Teams (Total Wickets)')
plt.xlabel('Total Wickets')
plt.ylabel('Teams')
plt.xticks(rotation=45)
plt.show()

# We can see both the tables and visual representation showing the analysis of top bowling teams. 
# Both of those clearly shows that Mumbai Indians are at the top place with highest number of wickets.
```

```python
# Data Analysis: Top 5 Bowlers from All the Seasons

# Counting dismissals for each bowler
top_bowlers = Data[Data['Wicket type'].notna()].groupby('Bowler').size().sort_values(ascending=False).head(5)

print(top_bowlers)

# Plotting top 5 bowlers as a line chart
plt.figure(figsize=(12, 6))
top_bowlers.plot(kind='line', color='green', marker='o')
plt.title('Top 5 Bowlers Across All Seasons (Total Wickets)')
plt.xlabel('Bowlers')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.show()

# We can see both the tables and visual representation showing the analysis of top 5 bowlers across all seasons. 
# Both of those clearly shows that R Ashwin at the top place with highest number of wickets.
```

```python
# Data Analysis: Top 5 Batsmen from All the Seasons

# Calculating total runs scored by each batsman
top_batsmen = Data.groupby('Batsman')['Runs_off_bat'].sum().sort_values(ascending=False).head(5)

print(top_batsmen)

# Plotting top 5 batsmen
plt.figure(figsize=(12, 6))
top_batsmen.plot(kind='line', color='purple')
plt.title('Top 5 Batsmen Across All Seasons (Total Runs)')
plt.xlabel('Batsmen')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()

# We can see both the tables and visual representation showing the analysis of top 5 batsman across all seasons. 
# Both of those clearly shows that V Kohli at the top place with highest number of runs.
```

```python
# Data Analysis: Most Common Wicket Types

# Counting occurrences of each wicket type
wicket_types = Data['Wicket type'].value_counts()

# Exclude 'None' from the data
wicket_types = wicket_types[wicket_types.index != 'None']

print(wicket_types)

# Plotting most common wicket types
plt.figure(figsize=(12, 6))
wicket_types.plot(kind='bar', color='red')
plt.title('Most Common Wicket Types')
plt.xlabel('Wicket Types')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# We can see both the tables and visual representation showing the analysis of most common wicket types across all seasons. 
# Both of those clearly shows caught is the most dismissal type followed by bowled.
```

```python
# Data Analysis: Players with Most Player of the Match Awards
# Counting player of the match awards for each player

player_of_match_counts = Data[Data['Ball'] == 0.1][Data['Innings']==1]
Man_of_the_match = player_of_match_counts['Player of match'].value_counts()

print(Man_of_the_match)

# Plotting players with most player of the match awards
plt.figure(figsize=(12, 6))
Man_of_the_match.plot(kind='line', color='gold')
plt.title('Players with Most Player of the Match Awards')
plt.xlabel('Players')
plt.ylabel('Number of Awards')
plt.xticks(rotation=45)
plt.show()

# We can see both the tables and visual representation showing the analysis of playerrs with most player of match awards across all seasons. 
# Both of those clearly shows AB de Villiers is at the top followed by SL Malinga
```

```python
# Data Analysis: Top Toss Winning Teams

# Counting toss wins for each team
toss_wins = Data['Toss winner'].value_counts().head(10)

print(toss_wins)

# Plotting top toss winning teams
plt.figure(figsize=(12, 6))
toss_wins.plot(kind='bar', color='blue')
plt.title('Top Toss Winning Teams')
plt.xlabel('Teams')
plt.ylabel('Number of Toss Wins')
plt.xticks(rotation=45)
plt.show()

# We can see both the tables and visual representation showing the analysis of playerrs with most top toss winning teams. 
# Both of those clearly shows Mumbai Indians is at the top followed by Chennai Super Kings.
```

```python
# Data Analysis: Evaluate Team Success Under Different Conditions (Batting First vs Chasing)

# Creating a new column to identify if the winning team batted first or chased
Data['win_condition'] = Data.apply(lambda x: 'Batting First' if (x['Toss winner'] == x['Winner'] and x['Toss decision'] == 'bat') or (x['Toss winner'] != x['Winner'] and x['Toss decision'] == 'field') else 'Chasing', axis=1)

# Counting the number of wins under each condition
win_conditions = Data['win_condition'].value_counts()

print(win_conditions)

# Plotting team success under different conditions
plt.figure(figsize=(12, 6))
win_conditions.plot(kind='bar', color=['green', 'orange'])
plt.title('Team Success Under Different Conditions (Batting First vs Chasing)')
plt.xlabel('Win Condition')
plt.ylabel('Number of Wins')
plt.xticks(rotation=0)
plt.show()

# We can see both the tables and visual representation showing the analysis of Team success under different conditions(batting first vs chasing). 
# Most of the teams won when they have chased in the second innings.
```

```python
# Data Analysis: Winner on Home Ground vs Away

# Creating a new column to identify if the winning team played on their home ground
# Assuming 'team1' is the home team for simplicity (you can adjust based on the actual data)
Data['home_ground_win'] = Data.apply(lambda x: 'Home Win' if x['Winner'] == x['Batting Team'] else 'Away Win', axis=1)

# Counting the number of home and away wins
home_away_wins = Data['home_ground_win'].value_counts()

print(home_away_wins)

# Plotting the number of wins on home ground vs away
plt.figure(figsize=(12, 6))
home_away_wins.plot(kind='bar', color=['purple', 'brown'])
plt.title('Wins on Home Ground vs Away')
plt.xlabel('Win Type')
plt.ylabel('Number of Wins')
plt.xticks(rotation=0)
plt.show()

# We can see both the tables and visual representation showing the analysis of wins on home ground vs away. 
# It's clear that away win are more compared to home win.
```

```python
# Data Analysis: Toss Decision Success Rate

# Creating a new column to identify if the toss decision led to a win
Data['toss_success'] = Data.apply(lambda x: 'Successful' if x['Toss winner'] == x['Winner'] else 'Unsuccessful', axis=1)

# Counting the success and failure of toss decisions
toss_success_rate = Data['toss_success'].value_counts()

print(toss_success_rate)

# Plotting the toss decision success rate
plt.figure(figsize=(12, 6))
toss_success_rate.plot(kind='bar', color=['green', 'red'])
plt.title('Toss Decision Success Rate')
plt.xlabel('Toss Decision Outcome')
plt.ylabel('Number of Matches')
plt.xticks(rotation=0)
plt.show()

# We can see both the tables and visual representation showing the analysis of Toss decision success rate. 
# It's obvious that toss decision played crucial role and there is slightly good chance of winning a match if they win the toss.
```

```python
most_productive_teams = Data.groupby('Batting Team')['Runs_off_bat'].sum().sort_values(ascending=False)

# Import Seaborn
import seaborn as sns

# Convert the data into a DataFrame suitable for a heatmap
most_productive_teams_df = most_productive_teams.reset_index()
most_productive_teams_df.columns = ['Teams', 'Total Runs']

print(most_productive_teams_df)

# Pivot the DataFrame to make it suitable for heatmap plotting
heatmap_data = most_productive_teams_df.set_index('Teams').T

# Plotting most productive teams using Seaborn heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, cmap='Blues', fmt='g')

# Adding title and labels
plt.title('Most Productive Teams (Total Runs)')
plt.xlabel('Teams')
plt.ylabel('Total Runs')

# Show the plot
plt.show()

# We can see both the tables and visual representation showing the analysis of most productive teams. 
# Mumbai indians tops list with 37524 runs followed by Royal challengers bangalore with 35810.
```

```python
# Data Analysis: Contribution of Extras to the Score

# Calculating total runs and extras for each team
total_runs = Data.groupby('Batting Team')['Runs_off_bat'].sum()
total_extras = Data.groupby('Batting Team')['Extras'].sum()

# Calculating the percentage contribution of extras
total_contribution = (total_extras / (total_runs + total_extras)) * 100

print(total_contribution)

# Plotting the contribution of extras to the score
plt.figure(figsize=(12, 6))
total_contribution.plot(kind='bar', color='red')
plt.title('Percentage Contribution of Extras to the Score')
plt.xlabel('Teams')
plt.ylabel('Percentage of Extras')
plt.xticks(rotation=45)
plt.show()

# We can see both the tables and visual representation showing the analysis of the percentage contribution of extras to the score. 
# Kings XI Punjab tops the list which implies low performance in the bowling. 
```
